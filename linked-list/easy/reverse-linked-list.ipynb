# Reverse Linked List

Given the `head` of a singly linked list, reverse the list, and return the reversed list.

**Example 1:**



```py
Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]
```

**Example 2:**



```py
Input: head = [1,2]
Output: [2,1]
Example 3:

Input: head = []
Output: []
```

**Constraints:**

- The number of nodes in the list is the range `[0, 5000]`.
- `-5000 <= Node.val <= 5000`

**Follow up:** A linked list can be reversed either iteratively or recursively. Could you implement both?


1 -> 2 -> 3 -> None\
3 -> 2 -> 1 -> None


````python
from typing import Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __repr__(self) -> str:
        return str(self.val)


def reverseList(head: Optional[ListNode]) -> Optional[ListNode]:
    a, b = None, head

    while b:
        c = b.next
        # print(a, b, c)
        b.next = a
        a = b
        b = c
    return a


c = ListNode(3)
b = ListNode(2, c)
a = ListNode(1, b)

reverseList(a)
````

## Validation


````python
c = ListNode(3)
b = ListNode(2, c)
a = ListNode(1, b)
h = reverseList(a)
h.next, h.next.next, h.next.next.next
print(h.next == b and h.next.next == a and h.next.next.next is None)

a = ListNode(1)
h = reverseList(a)
h.next == None
````
